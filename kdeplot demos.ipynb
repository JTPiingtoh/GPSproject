```python
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import norm 
import scipy.stats
import statistics 
```

```python
# Load the dataset and prepare the x-axis data
penguins = sns.load_dataset("penguins")
x_axis = penguins["bill_length_mm"].dropna()[:6]
x_axis -= 36.7
```

```python
sns.displot(x_axis, kde=True)
```

```python
# KDE plot with a defined bandwidth
sns.displot(x_axis, kde=True)

kde = scipy.stats.gaussian_kde(x_axis)
bandwidth = 1.06 * np.std(x_axis) * kde.scotts_factor()

# Overlay individual normal distributions
for x in x_axis:
    values = np.arange(x - 3, x + 3, 0.01)
    plt.plot(values, norm.pdf(values, x, bandwidth) / len(x_axis), 
             color='red')

# Configure plot limits and labels
plt.ylim([0, 2])
plt.legend()
plt.title("KDE with Individual Kernel Contributions")
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


# KDE plot with a defined bandwidth
sns.kdeplot(x_axis, bw_adjust=1, color='blue', label='KDE', bw_method="scott")

kde = scipy.stats.gaussian_kde(x_axis)
bandwidth = 1.06 * np.std(x_axis) * kde.scotts_factor()

# Overlay individual normal distributions
for x in x_axis:
    values = np.arange(x - 3, x + 3, 0.01)
    plt.plot(values, norm.pdf(values, x, bandwidth) / len(x_axis), 
             color='red')

# Configure plot limits and labels
plt.ylim([0, 0.4])
plt.legend()
plt.title("KDE with Individual Kernel Contributions")
plt.show()
```
